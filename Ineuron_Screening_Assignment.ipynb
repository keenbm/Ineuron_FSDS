```python
def read_replace(file_path,replacement_string):
  with open(file_path, "r") as file:
    original_content=file.read()
    new_content=original_content.replace("placement",replacement_string)
  with open(file_path, "w") as file:
    new_content=original_content.replace("placement",replacement_string)
    file.write(original_content)
  return (original_content,new_content)
```

```python
## Abstract Class Demo. and Decorator Demo.

import abc
from abc import ABC, abstractmethod
## Pyhon does not have support ABSTRACT class Natively so importing abc library


class abstract_Class(ABC):
  @abstractmethod ## By this decorator , we can defind abstract method 
  def color(self):
    pass

class apple(abstract_Class):
  def color(self):
    print("I am Red")

class watermelon(abstract_Class):
  def color(self):
    print("I am Green")

App=apple()
App.color()

App=watermelon()
App.color()
```

```python
# Multiple Inheritance 

class CMD():
    def Magane(self):
        print("I am managing entire company")
        
class Employee():
    def work(self):
        print("We do work")

class HR(CMD, Employee):
    def people(self):
        print("We manage everyone")
output = HR()
output.people()
output.work()
output.Magane()
```
